## 순위 선정

```python
c_y = pd.merge(pkl_용산구, df, how='outer')
c_y = c_y.sort_values(by=['동'])
c_y
```

```python
c_y = c_y.dropna(0)
c_y
```

```python
c_y['result'] = c_y['mean'] / c_y['주차장']
#c_y['result'] = c_y.sort_values(by=['result'], ascending=False)
c_y
```
