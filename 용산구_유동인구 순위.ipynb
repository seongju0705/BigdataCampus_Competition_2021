# 용산구_유동인구 순위
* 용산구_유동인구_비율.csv 이용
* 서울시 주차장(동별) 통계.xls 이용
* 유동인구 * 10000 / 주차장 수 -> 순위 선정

```python
import pandas as pd
import numpy as np
```

## 용산구_주차장 데이터

```python
pk = pd.read_excel("data/서울시 주차장(동별) 통계.xls")
```

```python
columns = ["자치구", "동", "합계.1"]
pk = pk[columns].copy()
pk = pk.drop([0])
```

```python
# 용산구
is_y = pk['자치구'] == '용산구'
pk_y = pk[is_y]
pk_y
```

```python
#컬럼 이름 변경
pk_y.rename(columns = {'합계.1' : '주차장'}, inplace = True)
pk_y
```

## 유동인구 데이터

```python
p = pd.read_csv("data/용산구_유동인구_비율.csv")
p
```

```python
p['P'] = p['P'] * 10000
p
```

```python
# 컬럼 이름 변경
p.rename(columns = {'ADMI_NM' : '동'}, inplace = True)
p
```

## 순위 선정

```python
#데이터 병합
df = pd.merge(pk_y, p, how='outer')
df = df.sort_values(by=['동'])
del df['Unnamed: 0']

#Null 값 제거
df = df.dropna(0)
df
```

```python
df['result'] = df['P'] / df['주차장']
df = df.sort_values(by=['result'], ascending=False)
df
```

```python
df.to_csv("용산구_유동인구 순위.csv")
```
